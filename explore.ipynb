```python
import pandas as pd
label = pd.read_pickle('processed_data/labels.pkl')
```

```python
all_primary_label = []
for infodict in label:
    if 'primary_labels' in infodict.keys():
        item = infodict['primary_labels']
    else:
        item = infodict['primary_label']
    if isinstance(item,list):
        print(item)
        all_primary_label.extend(item)
    else:
        all_primary_label.append(item)
```
